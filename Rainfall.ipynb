## **Data Visualization**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig=plt.figure(figsize=[6,6])
ax=fig.add_axes([0,0,1,1])
ax.set_title('Rainfall')
ax.boxplot(df['Jan-Feb'])
fig.show()
```

```python
rain_list=[df['Jan-Feb'],df['Mar-May']]
fig=plt.figure(figsize=[6,5])
ax=fig.add_axes([0,0,1,1])
ax.set_title('Rainfall')
ax.boxplot(rain_list,widths=0.5)
fig.show()
```

```python
fig=plt.figure(figsize=[10,5])
ax=fig.add_axes([0,0,1,1])
ax.scatter(df['Jan-Feb'],df['Mar-May'])
ax.set_title('Rainfall')
fig.show()
```

```python
fig=plt.figure(figsize=[7,5])
ax=fig.add_axes([0,0,1,1])
ax.set_title('Rainfall')
ax.bar(df['Mar-May'],df['Jun-Sep'])
fig.show()
```

```python
fig=plt.figure(figsize=[8,6])
ax=fig.add_axes([0,0,1,1])
ax.hist(df['JAN'])
ax.set_title('Rainfall')
fig.show()
```

```python
fig=plt.figure(figsize=[8,6])
ax=fig.add_axes([0,0,1,1])
ax.set_title('Rainfall')
ax.pie(df['JAN'],labels=df['DISTRICT'])
fig.show()
```
